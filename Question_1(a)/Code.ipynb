```python
fuel_data = data[["ENGINESIZE", "CYLINDERS", "FUELCONSUMPTION_COMB", "CO2EMISSIONS"]]
```

```python
X_fuel = fuel_data[["ENGINESIZE", "CYLINDERS", "FUELCONSUMPTION_COMB"]]
y_fuel = fuel_data["CO2EMISSIONS"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_fuel, y_fuel, test_size=0.2, random_state=42)
```

```python
#Linear Regression

from sklearn.linear_model import LinearRegression

lr_fuel = LinearRegression()
lr_fuel.fit(X_train, y_train)
```

```python
y_pred_fuel = lr_fuel.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

R2 = r2_score(y_test, y_pred_fuel)
MAE = mean_absolute_error(y_test, y_pred_fuel)
MSE = mean_squared_error(y_test, y_pred_fuel)
RMSE = np.sqrt(MSE)

print("CO2 Emission Model Performance")
print(f"R2 Score: ", R2)
print(f"MAE: ", MAE)
print(f"MSE: ", MSE)
print(f"RMSE: ", RMSE)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
sns.scatterplot(x=y_test, y=y_pred_fuel, color='blue', edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title("CO2 Emission Prediction: Actual vs Predicted")
plt.xlabel("Actual CO2 Emission")
plt.ylabel("Predicted CO2 Emission")
plt.grid(True)
plt.tight_layout()
plt.show()
```
