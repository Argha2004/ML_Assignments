```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
X = X.apply(pd.to_numeric, errors='coerce')

X = X.fillna(X.mean())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

results = {}
roc_curves = {}
```

```python
from sklearn.svm import SVC

for kernel in kernels:
    print(f"\nTraining SVM with {kernel} kernel...")
    model = SVC(kernel=kernel, probability=True, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
```

```python
if y.dtype == 'O':
        pos_label = 'M'
        y_test_mapped = y_test.map({'B': 0, 'M': 1})
else:
        pos_label = 1
        y_test_mapped = y_test
```

```python
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, f1_score,
    jaccard_score, confusion_matrix, roc_curve, auc
)

pos_label = 4

accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, pos_label=pos_label)
precision = precision_score(y_test, y_pred, pos_label=pos_label)
f1 = f1_score(y_test, y_pred, pos_label=pos_label)
jaccard = jaccard_score(y_test, y_pred, pos_label=pos_label)
error_rate = 1 - accuracy
cm = confusion_matrix(y_test, y_pred)

pos_label = 1

results[kernel] = {
      'Accuracy': accuracy,
      'Recall': recall,
      'Precision': precision,
      'F1-Score': f1,
      'Jaccard': jaccard,
      'Error Rate': error_rate,
      'Confusion Matrix': cm
}

```

```python
if y_test_mapped.dtype == 'O':
    y_test_binary = y_test_mapped.map({'B': 0, 'M': 1})
else:
    unique_vals = sorted(y_test_mapped.unique())
    y_test_binary = y_test_mapped.replace({unique_vals[0]: 0, unique_vals[1]: 1})

fpr, tpr, _ = roc_curve(y_test_binary, y_prob)
roc_auc = auc(fpr, tpr)
roc_curves[kernel] = (fpr, tpr, roc_auc)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f'Confusion Matrix ({kernel} Kernel)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print("\n=== Model Evaluation Results ===")
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
plt.figure(figsize=(8,6))
for kernel, (fpr, tpr, roc_auc) in roc_curves.items():
    plt.plot(fpr, tpr, label=f'{kernel} (AUC = {roc_auc:.2f})')

plt.plot([0,1], [0,1], 'k--')
plt.title('ROC Curve Comparison for SVM Kernels')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid(True)
plt.show()
```
