```python
#Add Proper Column Names

data.columns = [
    "Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
    "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"
]
```

```python
X = data.drop("Outcome", axis=1)
y = data["Outcome"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score, f1_score, classification_report

gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)

accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
f1_gnb = f1_score(y_test, y_pred_gnb)
print("\nGaussian Naive Bayes Results:")
print(f"Accuracy: {accuracy_gnb:.2f}")
print(f"F1 Score: {f1_gnb:.2f}")
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
y_pred_mnb = mnb.predict(X_test)

accuracy_mnb = accuracy_score(y_test, y_pred_mnb)
f1_mnb = f1_score(y_test, y_pred_mnb)
print("\nMultinomial Naive Bayes Results:")
print(f"Accuracy: {accuracy_mnb:.2f}")
print(f"F1 Score: {f1_mnb:.2f}")
```

```python
print("\nClassification Report (GaussianNB):")
print(classification_report(y_test, y_pred_gnb))

print("Classification Report (MultinomialNB):")
print(classification_report(y_test, y_pred_mnb))
```
